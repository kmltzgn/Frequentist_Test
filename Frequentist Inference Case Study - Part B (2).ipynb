## Hospital medical charges

Imagine that a hospital has hired you as their data scientist. An administrator is working on the hospital's business operations plan and needs you to help them answer some business questions. 

In this assignment notebook, you're going to use frequentist statistical inference on a data sample to answer the questions:
* has the hospital's revenue stream fallen below a key threshold?
* are patients with insurance really charged different amounts than those without?

Answering that last question with a frequentist approach makes some assumptions, and requires some knowledge, about the two groups.

We are going to use some data on medical charges obtained from [Kaggle](https://www.kaggle.com/easonlai/sample-insurance-claim-prediction-dataset). 

For the purposes of this exercise, assume the observations are the result of random sampling from our single hospital. Recall that in the previous assignment, we introduced the Central Limit Theorem (CLT), and its consequence that the distributions of sample statistics approach a normal distribution as $n$ increases. The amazing thing about this is that it applies to the sampling distributions of statistics that have been calculated from even highly non-normal distributions of data! Recall, also, that hypothesis testing is very much based on making inferences about such sample statistics. You're going to rely heavily on the CLT to apply frequentist (parametric) tests to answer the questions in this notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from numpy.random import seed
medical = pd.read_csv('insurance2.csv')
```

__Q1:__ Plot the histogram of charges and calculate the mean and standard deviation. Comment on the appropriateness of these statistics for the data.

__A:__As can be seen from the histogram, as well as from the standard deviation, the data does not have the shape of normal distribution. 

```python
_ = plt.hist(medical['charges'], bins=20)
_ = plt.xlabel('Charges ($)')
_ = plt.ylabel('Counts')
_ = plt.title('Distribution of charges')
```

```python
mean_charge = np.mean(medical['charges'])
std_charge = np.std(medical['charges'])

print(mean_charge)
print(std_charge)
```

__Q2:__ The administrator is concerned that the actual average charge has fallen below 12,000, threatening the hospital's operational model. On the assumption that these data represent a random sample of charges, how would you justify that these data allow you to answer that question? And what would be the most appropriate frequentist test, of the ones discussed so far, to apply?

__A:__ Justification can be done as follows: our hypothesis is the average charge is greater than 12000, and for the alternative hypothesis we can set the mean charges <= 12000 and apply a one-sided t test. 

__Q3:__ Given the nature of the administrator's concern, what is the appropriate confidence interval in this case? A ***one-sided*** or ***two-sided*** interval? (Refresh your understanding of this concept on p. 399 of the *AoS*). Calculate the critical value and the relevant 95% confidence interval for the mean, and comment on whether the administrator should be concerned.

__A:__

```python
t_value = t.ppf(0.975, 1337)

print(t_value)
```

```python
confidence_interval_t = t_value * std_charge / np.sqrt(1338)

print(confidence_interval_t)
```

```python
confidence_interval_low_t = mean_charge - confidence_interval_t

confidence_interval_high_t = mean_charge + confidence_interval_t

print("Confidence interval for t distribution is in between", confidence_interval_low_t, "and", confidence_interval_high_t)

```

The administrator then wants to know whether people with insurance really are charged a different amount to those without.

__Q4:__ State the null and alternative hypothesis here. Use the _t_-test for the difference between means, where the pooled standard deviation of the two groups is given by:
\begin{equation}
s_p = \sqrt{\frac{(n_0 - 1)s^2_0 + (n_1 - 1)s^2_1}{n_0 + n_1 - 2}}
\end{equation}

and the *t*-test statistic is then given by:

\begin{equation}
t = \frac{\bar{x}_0 - \bar{x}_1}{s_p \sqrt{1/n_0 + 1/n_1}}.
\end{equation}

(If you need some reminding of the general definition of ***t-statistic***, check out the definition on p. 404 of *AoS*). 

What assumption about the variances of the two groups are we making here?

```python
mean_no_insurance = medical[medical["insuranceclaim"] == 0].charges.mean()
mean_insurance = medical[medical["insuranceclaim"] == 1].charges.mean()

print(mean_no_insurance)
print(mean_insurance)
```

```python
std_no_insurance = medical[medical["insuranceclaim"] == 0].charges.std()
std_insurance = medical[medical["insuranceclaim"] == 1].charges.std()

print(std_no_insurance)
print(std_insurance)
```

```python
n_0 = len(medical[medical["insuranceclaim"] == 0])

n_1 = len(medical[medical["insuranceclaim"] == 1])

std_pooled = np.sqrt(((n_0-1)*std_no_insurance**2+(n_1-1)*std_insurance**2) / ((n_0+n_1)-2))

t_manual = (mean_no_insurance - mean_insurance) / (std_pooled * np.sqrt(float(1/n_0) + float(1/n_1)))


print(std_pooled)
print(t_manual) 
    
```

Congratulations! Hopefully you got the exact same numerical results. This shows that you correctly calculated the numbers by hand. Secondly, you used the correct function and saw that it's much easier to use. All you need to do is pass your data to it.
